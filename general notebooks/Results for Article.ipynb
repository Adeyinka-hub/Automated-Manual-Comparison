```python
results = pd.DataFrame({'method': ['Manual', 'Featuretools'], 
                        'time': [10, 1], 
                        'n_features': [30, 1850], 
                        'score': [100 * ((0.786 - 0.745) / (0.808 - 0.745)), 
                                  100 * ((0.787 - 0.745) / (0.808 - 0.745))]})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
results = results.set_index('method')
```

```python
plt.style.use('fivethirtyeight')

results['time'].plot.bar(color = 'blue', figsize = (10, 8), edgecolor = 'k', linewidth = 2);
plt.xlabel('', size = 24); plt.ylabel('Hours', size = 28);
plt.title('Development Time', size = 32);
plt.yticks([0, 2, 4, 6, 8, 10, 12], size = 24)
plt.xticks([0,1], ['Manual', 'Featuretools'], size = 24, rotation = 0);
```

```python
results['n_features'].plot.bar(color = 'forestgreen', figsize = (10, 8), edgecolor = 'k', linewidth = 2);
plt.xlabel('', size = 24); plt.ylabel('Features', size = 28);
plt.title('Number of Features', size = 32);
plt.yticks(size = 24)
plt.xticks([0,1], ['Manual', 'Featuretools'], size = 24, rotation = 0);
```

```python
results['score'].plot.bar(color = 'darkred', figsize = (10, 8), edgecolor = 'k', linewidth = 2);
plt.xlabel('', size = 24); plt.ylabel('% Improvement', size = 28);
plt.title('Improvement Relative to Baseline', size = 32);
plt.yticks(size = 24)
plt.xticks([0,1], ['Manual', 'Featuretools'], size = 24, rotation = 0);
```
