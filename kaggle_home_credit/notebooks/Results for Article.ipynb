```python
import pandas as pd
```

```python
feature_matrix = pd.read_csv('../input/feature_matrix.csv', nrows = 5)
feature_matrix
```

```python
feature_matrix.sample(10, axis = 1)
```

```python
results = pd.DataFrame({'method': ['Manual', 'Featuretools'], 'time': [30, 1], 'n_features': [30, 1850], 'score': [100 * ((0.786 - 0.745) / (0.808 - 0.745)), 100 * ((0.787 - 0.745) / (0.808 - 0.745))]})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
results = results.set_index('method')
```

```python
results['time'].plot.bar(color = 'blue', figsize = (10, 8), edgecolor = 'k');
plt.xlabel('', size = 24); plt.ylabel('Hours', size = 24);
plt.title('Development Time', size = 28);
plt.yticks([0, 5, 10, 15, 20, 25, 30], size = 20)
plt.xticks([0,1], ['Manual', 'Featuretools'], size = 22, rotation = 0);
```

```python
results['n_features'].plot.bar(color = 'forestgreen', figsize = (10, 8), edgecolor = 'k');
plt.xlabel('', size = 24); plt.ylabel('Features', size = 24);
plt.title('Number of Features', size = 28);
plt.yticks(size = 20)
plt.xticks([0,1], ['Manual', 'Featuretools'], size = 22, rotation = 0);
```

```python
results['score'].plot.bar(color = 'darkred', figsize = (10, 8), edgecolor = 'k');
plt.xlabel('', size = 24); plt.ylabel('% Improvement', size = 24);
plt.title('Improvement Relative to Baseline', size = 28);
plt.yticks(size = 20)
plt.xticks([0,1], ['Manual', 'Featuretools'], size = 22, rotation = 0);
```
